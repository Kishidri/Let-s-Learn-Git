### Amplitude
Vertical distance from the top of the peak to the center line

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,4*np.pi, 0.1)
y1 = np.sin(x)
A = 5
y2 = A*np.sin(x)
plt.plot(x, y1)
plt.plot(x, y2)
```

### Period vs Frequency
Period: Amount of time to complete a cycle
Frequency: How many times an event repeats on time

Frequency = 1 / Period

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,4*np.pi, 0.1)
y1 = np.sin(x)
f = 5
y3 = A*np.sin(x*f)
plt.plot(x, y1)
plt.plot(x, y3)
```

### Phase Shift (delay)

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,4*np.pi, 0.1)
y1 = np.sin(x)
shift = (3/4) *np.pi
y4 = A*np.sin(x + shift)
plt.plot(x, y1, label='y1')
plt.plot(x, y4,label = 'y4')
plt.legend()
```
